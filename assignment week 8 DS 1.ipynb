```python
# Create a new column in the dataframe called AgeRange that categorises passengers
# into age ranges: "Child" for ages 0-12, "Teen" for ages 13-19, "Adult" for ages 20-59,
# and "Senior" for ages 60 and above.
```

```python
age_bins = [0, 12, 19, 59, 81]
age_labels = ["Child", "Teen", "Adult", "Senior"]
data['AgeRange'] = pd.cut(data.Age, bins=age_bins, labels=age_labels, right=False)
```

```python
# Create a bar chart that shows the total number of passengers in each age range.
```

```python
#import seaborn as sns
#import matplotlib.pyplot as plt
```

```python
data['AgeRange'] = pd.cut(data.Age, bins=age_bins, labels=age_labels, right=False)
gb = data.groupby('AgeRange').size()
sns.barplot(x = gb.index, y = gb.values)
plt.ylabel('No. of People')
plt.title('The Total Number of Passengers In Each Age Range')
plt.show()
# bar graph shows The Total Number of Passengers In Each Age Range with respect to number of travellers.
```

```python
# Use the pivot_table() method to create a pivot table that shows the survival rate of
# passengers based on their sex, class, and age range.
```

```python
pivot_table = data.pivot_table(index = ['Sex', 'Pclass','AgeRange'], values = "Survived", aggfunc = "mean")
```

```python
pivot_table
```

```python
# This pivot table shows the difference in reference to sex, class and age range with respect to their survival rate.
```

```python
# Create a scatter plot that shows the relationship between age and fare. Color the
# points based on whether the passenger survived or not.
```

```python
sns.set(style="whitegrid")

sns.scatterplot(data, x ="Age", y="Fare", hue="Survived", palette={1: "blue", 0: "red"}, s=100)


plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Relationship between Age and Fare (Colored by Survival)")
plt.legend(title="Survived", labels=["No", "Yes"])

# Show the plot
plt.show()
```
